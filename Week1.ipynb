```
def PatternCount(Text, Pattern):
  Pointer = 0
  for x in range(len(Text)-len(Pattern)+1):
    if Text[x:x+len(Pattern)] == Pattern:
      Pointer += 1
  return Pointer
  
```

```
def FrequentWords(Text, k):
  Pointer = {}
  
  for x in range(len(Text)-k+1):
    if Text[x:x+k] in Pointer.keys():
      Pointer[Text[x:x+k]] += 1
    else:
      Pointer[Text[x:x+k]] = 0
  t = pd.Series(Pointer)
  print(' '.join(t[t==t.max()].index.values.tolist()))
```

```
def ReverseComplement(Input):
  Change = {'A':'T','T':'A','C':'G','G':'C'}
  Pointer = ''.join([Change[x] for x in Input[::-1]])
  print(Pointer)
```

```
def ClumpFinding(Genome, k, L, t):
  dico = {}
  for Scan in range(len(Genome) - L+1):
    for x in range(len(Genome[Scan:Scan+L]) - k + 1):
      if str(Scan)+Genome[Scan:Scan+L][x:x+k] in dico.keys():
        dico[str(Scan)+Genome[Scan:Scan+L][x:x+k]] += 1
      else:
        dico[str(Scan)+Genome[Scan:Scan+L][x:x+k]] = 1
  if [pd.Series(dico) >= t] == True:
    print(pd.Series(dico)[pd.Series(dico) >= t].index.str[-5:].unique())
  else:
    print('There is none')

```

```
def PatternToNumber(Pattern):
  dico = {'A':0,'C':1,'G':2,'T':3}
  PatLen = len(Pattern)
  Number= 0
  for i in range(PatLen):
      Number += dico[Pattern[i]]*4**(PatLen-i-1)
  return Number
 
```
